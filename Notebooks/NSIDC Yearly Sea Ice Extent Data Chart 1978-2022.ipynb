```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../Dataset/NSIDC/N_Yearly_SeaIce_Extent_1978_2023.csv', names = ['Year', 'Extent'])
```

```python
df.head()
```

```python
df.tail(10)
```

```python
df = df.drop([45], axis=0)
```

```python
fig = plt.figure(figsize=(16, 8), facecolor='#000000')
ax = plt.subplot(facecolor='#000000')
ax.plot(df['Year'], df['Extent'], 'o-')
ax.set_xlabel('Year', fontsize=18)
ax.set_ylabel('Sea Ice Extent in million sq.km', fontsize=18)
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(colors='#ffffff', size=0)
ax.set_facecolor('#000000')
plt.title('Sea Ice Extent Distribution', color='#ffffff', fontsize=30)
plt.savefig('NSIDC_Yearly_Sea_Extent_Data_Chart_1978_2022.png')
plt.show()
```
